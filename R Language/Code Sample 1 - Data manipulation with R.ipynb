# European Values Survey - Years: 2008-2010

```r
evs <- haven::read_sav("European Values Survey 2008.sav", encoding = 'latin1')


### Creating Country Variables - turning numeric attributes to labels

att <- attr(evs$country, "labels")
att2 <- names(att[-2])
x <- as.factor(evs$country)
evs$pais <- factor(x, levels = levels(x), labels = att2)
```

```r
### Creating variables continent, survey year and question labels

evs$cont <- 'Europa'
evs$ano <- as.numeric(format(evs$v372,'%Y'))

# Inverting Trust variables Scales
evs$v221[evs$v221 == 4] <- 5 
evs$v221[evs$v221 == 3] <- 6 
evs$v221[evs$v221 == 2] <- 7 
evs$v221[evs$v221 == 1] <- 8 
evs$v221[evs$v221 == 5] <- 1 # none at all
evs$v221[evs$v221 == 6] <- 2 # not very much
evs$v221[evs$v221 == 7] <- 3 # quite a lot
evs$v221[evs$v221 == 8] <- 4 # a great deal

evs$v222[evs$v222 == 4] <- 5 
evs$v222[evs$v222 == 3] <- 6 
evs$v222[evs$v222 == 2] <- 7 
evs$v222[evs$v222 == 1] <- 8 
evs$v222[evs$v222 == 5] <- 1 # none at all
evs$v222[evs$v222 == 6] <- 2 # not very much
evs$v222[evs$v222 == 7] <- 3 # quite a lot
evs$v222[evs$v222 == 8] <- 4 # a great deal

evs$v211[evs$v211 == 4] <- 5 
evs$v211[evs$v211 == 3] <- 6 
evs$v211[evs$v211 == 2] <- 7 
evs$v211[evs$v211 == 1] <- 8 
evs$v211[evs$v211 == 5] <- 1 # none at all
evs$v211[evs$v211 == 6] <- 2 # not very much
evs$v211[evs$v211 == 7] <- 3 # quite a lot
evs$v211[evs$v211 == 8] <- 4 # a great deal


var_label(evs) <- list(v211 = 'Confiança no Parlamento',
                       v221 = 'Confiança nos Partidos',
                       v222 = 'Confiança no Governo',
                       cont = 'Continente',
                       pais = 'Países',
                       ano  = 'Ano da Entrevista',
                       v62  = 'Confiança nas outras pessoas',
                       v225 = 'Apoio a um líder forte')


### Creating sintetized database with country means

evs %>% select(pais, v221, v211, v222, ano, cont, v62, v225) %>% group_by(pais) %>% 
  rename(parl = v211, part = v221, gov = v222) %>% 
  summarise(part_mean = mean(part, na.rm = TRUE),
            parl_mean = mean(parl, na.rm = T),
            gov_mean  = mean(gov, na.rm = T),
            peoptrus  = (mean(v62, na.rm = T)-1)*10,
            autor     = mean(v225, na.rm = T)) -> evsum

rm("att", "att2", "x")
```

```r
# Collected variables: Partidarism

cses <- import("cses_imd.sav")

# Standardazing country names

cses$IMD1006_NAM[cses$IMD1006_NAM == "United States of America"] <- "United States"
cses$IMD1006_NAM[cses$IMD1006_NAM == "Great Britain"] <- "United Kingdom"
cses$IMD1006_NAM[cses$IMD1006_NAM == "Republic of Korea"] <- "South Korea"
cses$IMD1006_NAM[cses$IMD1006_NAM == "Slovakia"] <- "Slovak Republic"
cses$IMD1006_NAM[cses$IMD1006_NAM == "Russian Federation"] <- "Russia"

# PARTY IDENTIFICATION: ARE YOU CLOSE TO ANY POLITICAL PARTY?

# Standardizing to yes or no answers

cses$IMD3005_1[cses$IMD3005_1 == 7 |
                 cses$IMD3005_1 == 8 |
                 cses$IMD3005_1 == 9] <- NA

# - Renaming variables country and year
# - Filtering 2004-2010
# - grouping and summarizing mean for partidarism
# - Joining DALP selecting country and last year election variables
# - Ordering (grouped) lines by the difference between cses year and DALPs election year
# - Selecting countries in which years are the closest to DALPs


cses %>% rename("pais" = "IMD1006_NAM", "year" = "IMD1008_YEAR") %>% 
  filter(year == "2004" | year == "2005" | year == "2006" | year == "2007" |
         year == "2008" | year == "2009" | year == "2010") %>% 
  group_by(pais, year) %>% summarise(partida = mean(IMD3005_1, na.rm = T)) %>% 
  left_join(select(dalp, pais, ano_ultima_eleicao)) %>% 
  arrange(abs(as.numeric(ano_ultima_eleicao) - year)) %>%
  slice(1) %>% filter(ano_ultima_eleicao != is.na(ano_ultima_eleicao)) %>% 
  select(-ano_ultima_eleicao) -> cses

attr(cses$partida, "labels") <- c("No, not close to any party" = 0,
                                 "Yes, feels close to a party" = 1)
```

```r
World_Bank_GDP <- read_excel("World Bank GDP.xls")

#Setting 3 line as index
names(World_Bank_GDP) <- World_Bank_GDP[3,]

World_Bank_GDP[-1:-3,] %>% select("Country Name", contains("200")) -> pib
rm(World_Bank_GDP)

# Standardazing country names

pib$`Country Name`[pib$`Country Name` == "Egypt, Arab Rep."] <- "Egypt"
pib$`Country Name`[pib$`Country Name` == "Dominican Republic"] <- "República Dominicana"
pib$`Country Name`[pib$`Country Name` == "North Macedonia"] <- "Macedonia"
pib$`Country Name`[pib$`Country Name` == "Korea, Rep."] <- "South Korea"
pib$`Country Name`[pib$`Country Name` == "Russian Federation"] <- "Russia"
pib$`Country Name`[pib$`Country Name` == "Venezuela, RB"] <- "Venezuela"
pib$`Country Name`[pib$`Country Name` == "Egypt, Arab Rep."] <- "Egypt"

# Joining DALP, selecting GDP growth based on election years, in order:
# - Selecting country and year variables from DALP
# - Renaming variable country to the same name as in GDP database
# - Joins both databases
# - Removes NAs, necessary to accomplish next steps
# - Converts to data.table to be able to do the following steps
# - Selects the values from the only variable which has the same year as the variable
#'ano_ultima_eleicao'
# - Selecting interest variables
# - Renaming Country variable to integrate the datasets

dalp %>% select(pais, ano_ultima_eleicao) %>% 
  rename("Country Name" = "pais") %>% left_join(pib) %>%
filter(!is.na(ano_ultima_eleicao)) %>% data.table(.) %>% 
  .[, pib := .[, get(ano_ultima_eleicao), 1:nrow(.)]$V1] %>% 
  select(1, 13) %>% rename(., "pais" = `Country Name`) -> pib
```

# Inequality variables - Year 2004-2010 

```r

des <- import("swiid7_1_summary.csv")

# Standardazing country names

des$country[des$country == "Korea"] <- "South Korea"
des$country[des$country == "Slovakia"] <- "Slovak Republic"
des$country[des$country == "Dominican Republic"] <- "República Dominicana"

des %>% rename(pais = "country") %>% 
  filter(year == "2004" | year == "2005" | year == "2006" | year == "2007" |
           year == "2008" | year == "2009" | year == "2010") %>% group_by(pais) %>%
  left_join(select(dalp, pais, ano_ultima_eleicao)) %>% 
  filter(ano_ultima_eleicao != is.na(ano_ultima_eleicao)) %>% 
  arrange(abs(as.numeric(ano_ultima_eleicao) - year)) %>%
  slice(1) %>% select(pais, year, contains("gini")) -> des
```

# Creating final dataset: integrating all databases 

```r

dalp <- select(dalp, pais, party, e1, d4, sucesso, ano_ultima_eleicao, 
               sistemaeleitoral, sistemapolitico)

dalp %>% 
  left_join(wvs, by = "pais") %>% 
  left_join(gbs, by = "pais") %>% 
  left_join(cses, by = "pais") %>% 
  left_join(pib, "pais") %>% 
  left_join(idh, "pais") %>% 
  left_join(des, "pais") -> bancocompleto

# Joining 2 dinstinct variables from distinct datasets into one in
#the final database


# Trust in Party

bancocompleto$part_mean <- coalesce(bancocompleto$part_mean.x,
                                    bancocompleto$part_mean.y)
bancocompleto$part_mean_media <- (bancocompleto$part_mean.x +
                                  bancocompleto$part_mean.y)/2
bancocompleto$partconf <- coalesce(bancocompleto$part_mean_media,
                                   bancocompleto$part_mean)

# Trust in Parlament

bancocompleto$parl_mean <- coalesce(bancocompleto$parl_mean.x,
                                    bancocompleto$parl_mean.y)
bancocompleto$parl_mean_media <- (bancocompleto$parl_mean.x +
                                  bancocompleto$parl_mean.y)/2
bancocompleto$parlconf <- coalesce(bancocompleto$parl_mean_media,
                                   bancocompleto$parl_mean)

# Trust in Government

bancocompleto$gov_mean <- coalesce(bancocompleto$gov_mean.x,
                                   bancocompleto$gov_mean.y)
bancocompleto$gov_mean_media <- (bancocompleto$gov_mean.x +
                                 bancocompleto$gov_mean.y)/2
bancocompleto$govconf <- coalesce(bancocompleto$gov_mean_media,
                                  bancocompleto$gov_mean)


# Partidarism

bancocompleto$partida_mean <- coalesce(bancocompleto$partida.x,
                                       bancocompleto$partida.y)
bancocompleto$partmedia_mean <- (bancocompleto$partida.x +
                                 bancocompleto$partida.y)/2
bancocompleto$partida <- coalesce(bancocompleto$partmedia_mean,
                                  bancocompleto$partida_mean)

# Trust in People
bancocompleto$peoptrus <- coalesce(bancocompleto$peoptrus.x,
                                   bancocompleto$peoptrus.y)
bancocompleto$peoptrus_media <- (bancocompleto$peoptrus.x +
                                 bancocompleto$peoptrus.y)/2
bancocompleto$segu <- coalesce(bancocompleto$peoptrus_media,
                               bancocompleto$peoptrus)

# Autoritarism
bancocompleto$autor <- coalesce(bancocompleto$autor.x,
                                bancocompleto$autor.y)
bancocompleto$autor_media <- (bancocompleto$autor.x +
                              bancocompleto$autor.y)/2
bancocompleto$autori <- coalesce(bancocompleto$autor_media,
                                 bancocompleto$autor)


banc <- select(bancocompleto, -contains(".y"), -contains(".x"),
               -contains("mean")) %>% distinct()

# Creating Political Trust variable (mean between Trust in Parties,
# Parlament and Government

banc$conf <- (banc$partconf + banc$parlconf + banc$govconf)/3

banc <- select(banc, pais, party, ano_ultima_eleicao, e1:idh, contains("gini"),
               contains("conf"), partida, segu, autori)

# Removing problematic observation
banc <- filter(banc, !pais == "Egypt")

# Setting dependent variables attributes/ subtitles

banc$pib <- as.numeric(banc$pib)
banc$sistemaeleitoral <- factor(banc$sistemaeleitoral,
                                levels = c("Majoritario", "Ambos", "Proporcional"))
banc$sistemapolitico <- factor(banc$sistemapolitico,
                               levels = c("Parlamentar", "Misto", "Presidencial"))
banc$gini_mkt <- banc$gini_mkt/100

var_label(banc) <- list (pais = "Países",
                         party = "Partido do Executivo",
                         e1 = "Nível de apelo Carismático",
                         d4 = "Nível de apelo à Identidade Nacional",
                         sucesso = "Obteve ou não maior número de votos",
                         ano_ultima_eleicao = "Ano da Eleição",
                         sistemaeleitoral = "Sistema eleitoral da eleição",
                         sistemapolitico = "Sistema político da eleição",
                         partconf = "Confiança nos Partidos",
                         parlconf = "Confiança no Parlamento",
                         govconf = "Confiança no Governo",
                         partida = "Partidarismo",
                         conf = "Confiança Política",
                         pib = "Variação do PIB no ano eleitoral",
                         gini_mkt = "Índice de GINI pré-taxas",
                         idh = "Nível de desenvolvimento",
                         segu = "Confiança nas pessoas",
                         autori = "Grau de apoio a autoritarismo")

attr(banc$segu, "Question") <- "Generally speaking, would you say that most people
can be trusted or that you can’t be too careful in dealing with people?"

attr(banc$segu, "labels") <- c("Most people can be trusted" = 1,
                                  "Need to be very careful" = 10)       

attr(banc$autori, "Question") <- c("For each one, would you say it is a very good,
fairly good, fairly bad or very bad way of governing this country?", "Having a strong
leader who does not have to bother with parliament and elections")

attr(banc$autori, "labels") <- c("Very good" = 1,
                               "Fairly good" = 2,
                               "Fairly bad" = 3,
                               "Very bad" = 4)

attr(banc$gini_mkt, "labels") <- "0 corresponde à completa igualdade
(no caso do rendimento, por exemplo, toda a população recebe o mesmo salário)
e 1 corresponde à completa desigualdade (onde uma pessoa recebe todo o rendimento
e as demais nada recebem)."

attr(banc$sucesso, "labels") <- c("Venceu eleição (obteve maior número de votos)" = 1,
                                  "Não venceu a eleição (não obteve maior número de
                                  votos" = 0)

rm(list=setdiff(ls(), "banc"))

# export(banc, "banc.csv")
```
